```python
def create_csv_area_years(pollution: pd.DataFrame,year: str):
    df_year = pollution.loc[:,['area',year]].groupby(by='area').sum().sort_values(by=year ,ascending=False).reset_index().assign(ranking=lambda df: df.index + 1).reindex(columns=['ranking', 'area', year]).set_index('ranking').rename(columns={ year :'value'}).head(20)
    filename = 'data/CW' + str(year) + '.csv'
    df_year.to_csv(filename)

```
